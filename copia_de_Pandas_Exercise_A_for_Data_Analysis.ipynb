Selection Based on Conditionals

Let's see which stations have the most and fewest rides.


```python
# use the min() and max() modules
# your code here
import pandas as pd

# Cargar el dataset
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'
df = pd.read_csv(url)

# Supongamos que la columna que contiene el número de viajes se llama 'rides'
# Asegúrate de reemplazar 'rides' con el nombre real de la columna en tu dataset

# Encontrar la fila con el mayor número de viajes
max_rides_index = df['rides'].idxmax()
max_rides_station = df.loc[max_rides_index]

# Encontrar la fila con el menor número de viajes
min_rides_index = df['rides'].idxmin()
min_rides_station = df.loc[min_rides_index]

print("Estación con el mayor número de viajes:")
print(max_rides_station)

print("\nEstación con el menor número de viajes:")
print(min_rides_station)

pass
```

Create a Time Index

So we have all rides from January 1, 2001 to December 31, 2019. Let's do some statistics that group ride statistics by date. First we need to tell Pandas that the date column is in fact a date. So, we convert the column to a proper 'DateTime' type, and then set the index to this column.

This step takes a little bit of time!

The following picture show the desired output

image.png

```python
# hint, use to_datetime() and set_index()  modules
# your code here
import pandas as pd

# Cargar el dataset
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'
df = pd.read_csv(url)

# Convertir la columna de fecha a tipo 'DateTime'
# Asumamos que la columna de fecha se llama 'date'
df['date'] = pd.to_datetime(df['date'])

# Establecer la columna de fecha como el índice
df.set_index('date', inplace=True)

# Verificar la conversión y el nuevo índice
print(df.head())
print(df.index)

pass
```

```python
# Agrupar por año y calcular la suma de viajes
rides_per_year = df.resample('Y').sum()
print("\nNúmero total de viajes por año:")
print(rides_per_year)

# Agrupar por mes y calcular la suma de viajes
rides_per_month = df.resample('M').sum()
print("\nNúmero total de viajes por mes:")
print(rides_per_month)

# Agrupar por día y calcular la suma de viajes
rides_per_day = df.resample('D').sum()
print("\nNúmero total de viajes por día:")
print(rides_per_day)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo de Seaborn
sns.set(rc={'figure.figsize':(11, 4)})

# Cargar el dataset
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'
df = pd.read_csv(url)

# Asegúrate de que el nombre de la columna de la fecha y la estación es correcto
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)

# Filtrar datos para la estación 'Garfield-Dan Ryan'
garfield_dan_ryan = df[df['stationname'] == 'Garfield-Dan Ryan']

# Filtrar datos para la estación 'Garfield-South Elevated'
garfield_south_elevated = df[df['stationname'] == 'Garfield-South Elevated']

# Graficar los datos para 'Garfield-Dan Ryan'
plt.figure(figsize=(11, 4))
sns.lineplot(data=garfield_dan_ryan, x=garfield_dan_ryan.index, y='rides', linewidth=0.5)
plt.title('Número de Viajes en Garfield-Dan Ryan')
plt.xlabel('Fecha')
plt.ylabel('Número de Viajes')
plt.grid(True)
plt.show()

# Graficar los datos para 'Garfield-South Elevated'
plt.figure(figsize=(11, 4))
sns.lineplot(data=garfield_south_elevated, x=garfield_south_elevated.index, y='rides', linewidth=0.5)
plt.title('Número de Viajes en Garfield-South Elevated')
plt.xlabel('Fecha')
plt.ylabel('Número de Viajes')
plt.grid(True)
plt.show()
```
